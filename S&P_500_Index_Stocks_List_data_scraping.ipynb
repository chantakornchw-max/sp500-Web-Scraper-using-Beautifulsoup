# Scraping S&P 500 Index Stocks List 

In this project, I have implemented the data scraping pipeline using two distinct methodologies to demonstrate technical versatility in data
acquisition:

•Method 1: Utilized pd.read_html() for rapid table detection and conversion.

•Method 2: Create a manual loop to iterate through the th, tr, and td tags.

### Importing Core Libraries

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

•requests: Responsible for sending HTTP requests to the target URL and retrieving the raw HTML content from the server.

•BeautifulSoup: Used for parsing the raw HTML into a searchable "Soup" object, allowing for easy navigation through tags (like table, th, tr, and td).

•pandas: The primary data manipulation library used to transform the extracted lists into a structured DataFrame for cleaning and exporting.

•warnings: This library is used to manage system notifications. Calling filterwarnings('ignore') will suppress non-critical warning messages, ensuring that the notebook output remains clean and readable.

```python
url = 'https://stockanalysis.com/list/sp-500-stocks/'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36'
}

response = requests.get(url, headers=headers)
print(response)
```

•url: Specifies the target webpage containing the S&P 500 stock list.

•headers (User-Agent): This is a critical security bypass step. By providing a User-Agent string, the script identifies itself as a standard web browser (like Chrome) rather than a bot. This prevents the website from blocking the request due to automated scraping policies.

•requests.get(): Executes the actual retrieval of the page's HTML content using the defined URL and headers.

•<Response [200]>: The output confirms a successful connection. In HTTP status codes, 200 means the server processed the request perfectly and is ready to deliver the data.

```python
soup = BeautifulSoup(response.text, 'html')
print(soup)
```

•response.text: This is the raw HTML code of the webpage, which currently looks like a massive, unstructured string of characters.

•BeautifulSoup(..., 'html'): This function takes that raw string and "parses" it using an HTML engine. It organizes the code into a hierarchy, much like a family tree (the Document Object Model), where every button, table, and link is a specific "branch".

```python
table = soup.find_all('table')
print(table)
```

•soup.find_all('table'): This function searches through the entire organized "Soup" object for every occurrence of the table tag. In web development, tables are the standard container used to store structured lists, such as the S&P 500 stock index.

•table (The variable): The result is a Python List containing the HTML code for every table found on the page. By isolating these tags, we can ignore thousands of lines of irrelevant code

### Method 1: pd.read_html()

```python
df = pd.read_html(str(table))[0]
```

•str(table): Converts the BeautifulSoup object (which is a Python-specific object) back into a raw HTML string. This is necessary because the Pandas function requires a string or file-like object to process the data.

•pd.read_html(): This is a high-level Pandas function that acts as an "automated table detector". It scans the HTML string specifically for table, tr, and td tags, and automatically transforms them into a structured format.

### Method 2: Manual Loop 

```python
columns = [title.text.strip() for title in soup.find_all('th')]
print(columns)
```

•soup.find_all('th'): The script targets all <th> (Table Header) tags within the parsed HTML. In professional web design, these tags are the designated containers for header text.

•List Comprehension: This concise Pythonic method iterates through the found header tags to build a list of titles in a single, efficient line of code.

•.text.strip(): This is a crucial Data Cleansing step. The .text attribute extracts the raw content, while .strip() removes invisible characters like newlines (\n) or extra spaces that often persist in raw web data.

```python
column_data = soup.find_all('tr')
print(column_data)
```

•soup.find_all('tr'): The script searches the "Soup" object for all <tr> (Table Row) tags. In an HTML table, each <tr> represents one complete horizontal record—in this case, one company's worth of data.

•column_data: This variable now stores a list of these row elements. By capturing all rows first, we prepare the script to iterate through them one by one in the upcoming data-cleaning loop.

```python
all_rows = [] 

for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row = [data.text.strip() for data in row_data]
    
    all_rows.append(individual_row)
```

•for row in column_data[1:]: We initiate a loop starting from the second element (index 1). Slicing the list this way is essential to skip the header row (th).

•row.find_all('td'): Within each specific row, the script locates all <td> (Table Data) tags, which contain the actual stock values like price and market cap.

•List Comprehension with .strip(): For every row, we create a sub-list by extracting the .text from each cell and immediately applying .strip() to ensure no dirty HTML characters or extra spaces enter our dataset.

•all_rows.append(): Each completed sub-list (representing one company) is added to the all_rows master list, creating a 2D array structure ready for conversion.

```python
df = pd.DataFrame(all_rows, columns=columns)
```

•pd.DataFrame(): This is the core constructor used to build a labeled, two-dimensional data structure.

•all_rows: This serves as the "Body" of the dataset, containing the 500+ individual stock records extracted during the manual loop.

•columns=columns: This assigns the schema we extracted in colums variable, ensuring every column—from 'No.' to 'Revenue'—is accurately labeled.
